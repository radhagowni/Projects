```python
import random
original_num=random.randint(100,999)
original_num=str(original_num)
#print(original_num)
while True:
    guess_num=input("enter number:")
    if len(original_num)!=len(guess_num):
        print("enter a valid number")
        continue
    if len(original_num)!=len(set(guess_num)):
        print("duplicate number")
        continue
    if original_num==guess_num:
        print("fermi fermi fermi")
        print("Congratulations!,YOU WON")
        break
    else:
        output=[]
        for i in range(len(original_num)):
            for j in range(len(guess_num)):
                if original_num[i]==guess_num[j]:
                    if i==j:
                        output.append("fermi")
                        continue
                    else:
                        output.append("pico")
                        continue
        #print(output)
        if output==[]:
          print("bagel")
        else:
            out=""
            for i in output:
                    out=out+" "+i
            print(out)
            continue
```
